```python
!/Applications/bioinfo/bowtie2-2.4.2-macos-x86_64/bowtie2-build \
Danio_rerio.GRCz11.dna_sm.toplevel.fa.gz \
Dr_bowtie-index
```

### The above code uses bowtie2 to build an index of the reference genome of Danio rerio. 
### Once the index is built, take it and the trimmed files to get a SAM alignment file. 

```python
!/Applications/bioinfo/bowtie2-2.4.2-macos-x86_64/bowtie2 \
-x Dr_bowtie-index \
-1 ../analyses/trimmed/HIPPLAT_NEF95054_S2_R1.fq -2 ../analyses/trimmed/HIPPLAT_NEF95054_S2_R2.fq \
-p 4 \
--very-fast \
-S ../analyses/HIPP-DR.sam
```
